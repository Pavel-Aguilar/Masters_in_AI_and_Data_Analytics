```python
def separador():
    print("\n" + "=" * 170 + "\n")
def header(Titulo):
  por=((170-len(Titulo))//10)
  a=int(1.2*por)
  print("."*a+"·"*a+"~"*a+"≈"*a+"≋"
  *int(por*.2),Titulo,"≋"*int(por*.2)
  +"≈"*a+"~"*a+"·"*a+"."*a,"\n\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
import scipy.stats as stats

header("Descripción de los datos")
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data"
column_names = ['age', 'year', 'nodes', 'status']
df = pd.read_csv(url, header=None, names=column_names)

for col in df.select_dtypes(include="object").columns:
  df[col]=pd.to_numeric(df[col], errors="coerce")
print(f'Encabezado de la base de datos:\n{df.head(5)}')
separador()
print(f'Tamaño de la base de datos:\n{df.shape[0]} filas y {df.shape[1]} columnas')
separador()
print(f'Tipo de dato de cada columna:\n{df.info()}')
separador()
print(f'El EDA de los datos es:\n{df.describe()}')
separador()
print(f'La moda de cada columna es:\n{df.mode().iloc[0]}')
separador()
print(df.dtypes)
separador()

print('Nan\n', df.isna().sum())

mascara=df.isna().all(axis=1)
filasnan=mascara.sum()
print(f'Hay {filasnan} filas con todos los valores nulos')
print(df[mascara].head())

header("Histogramas para las columnas numéricas")
columnas = df.select_dtypes(include=['float64', 'int64']).columns
n_columnas=len(columnas)
n_filas=(n_columnas+2)//3

plt.figure(figsize=(15, 5*n_filas))
for i, col in enumerate(columnas):
    plt.subplot(n_filas, 3, i+1)
    sns.histplot(df[col], kde=True, bins=30, color='darkgreen')
    plt.title(f'Histograma de {col}', fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.ylabel('Frecuencia', fontsize=10)

plt.tight_layout()
plt.show()

header('Grafica de barras de la columna Status')
sns.set(style="whitegrid")
plt.figure(figsize=(15, 5))
ax=sns.countplot(x='status', data=df, palette='viridis')
plt.title('Distribución de la columna "Status"', fontsize=16, fontweight='bold')
plt.xlabel('Estado de Supervivencia', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
leyendas=['Sobrevivió mas de 5 años', 'Murió en 5 años o menos']
ax.set_xticklabels(leyendas)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

header('Boxplots de las columnas numéricas')
columnas=df.select_dtypes(include=['float64', 'int64']).columns
n_columnas=len(columnas)
n_filas=(n_columnas+2)//3

plt.figure(figsize=(15,5*n_filas))
for i, col in enumerate(columnas):
  ax=plt.subplot(n_filas, 3, i+1)
  bp=ax.boxplot(df[col].dropna(), patch_artist=True)
  for box in bp['boxes']:
    box.set(color='black', linewidth=1.5)
    box.set(facecolor='black', linewidth=1.5)
  for whisker in bp['whiskers']:
    whisker.set(color='black', linewidth=1.5)
  for median in bp['medians']:
    median.set(color='darkgreen', linewidth=1.5)
  for cap in bp['caps']:
    cap.set(color='black', linewidth=1.5)
  for flier in bp['fliers']:
    flier.set(marker='o', color='red', markersize=5, alpha=0.5)

  plt.title(f'Boxplot de {col}', fontsize=12)
  plt.xlabel('Distribución de valores', fontsize=10)
  ax.set_xticks([])
plt.tight_layout()
plt.show()

header('Matriz de correlación y mapa de calor')
columnas2=df.select_dtypes(include=['float64', 'int64'])
corr_matrix=columnas2.corr()
sns.set(style='white')
plt.figure(figsize=(10,12))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5, fmt='.2f', annot_kws={'size':8})
plt.title('Matriz de correlación', size=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

header('Pairplot para visualizar relaciones entre variables numéricas')
sns.pairplot(df, vars=columnas2.columns, diag_kind='kde')
#plt.suptitle('Pairplot para visualizar relaciones entre variables numéricas', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

header('Pruebas de normalidad')
datos=df.select_dtypes(include=[np.number]).dropna()
# Función para realizar pruebas de normalidad
def normality_tests(data, column_name):
    print(f"\nPruebas de normalidad para la columna: {column_name}")

    # Prueba de Shapiro-Wilk
    shapiro_test = stats.shapiro(data.sample(min(5000, len(data)), random_state=42))  # Muestra de 5000 para evitar problemas de cómputo
    print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")

    # Prueba de Kolmogorov-Smirnov
    ks_test = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    print(f"Kolmogorov-Smirnov Test: p-valor = {ks_test.pvalue:.5f}")

    # Prueba de Anderson-Darling
    anderson_test = stats.anderson(data, dist='norm')
    print(f"Anderson-Darling Test: Estadístico = {anderson_test.statistic:.5f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        if anderson_test.statistic < cv:
            print(f"  Al nivel de significancia {sl}%, los datos parecen normales.")
        else:
            print(f"  Al nivel de significancia {sl}%, los datos NO parecen normales.")

    # Prueba de D'Agostino-Pearson
    dagostino_test = stats.normaltest(data)
    print(f"D'Agostino-Pearson Test: p-valor = {dagostino_test.pvalue:.5f}")

# Función para graficar histograma y QQ-Plot
def plot_distribution(data, column_name):
    plt.figure(figsize=(12, 5))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=50, kde=True, color="skyblue")
    plt.title(f"Distribución de {column_name}")

    # QQ-Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"QQ-Plot de {column_name}")

    plt.show()

# Aplicar pruebas de normalidad y gráficos a cada columna numérica
for column in datos:
    data = df[column]
    plot_distribution(data, column)
    normality_tests(data, column)
```
