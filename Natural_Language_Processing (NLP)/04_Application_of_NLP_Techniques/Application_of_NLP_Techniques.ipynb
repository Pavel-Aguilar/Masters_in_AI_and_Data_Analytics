```python
#Se importan bibliotecas
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import spacy
nlp = spacy.load("es_core_news_sm")
import unicodedata
import re
```

```python
#Se descargan los recursos para los stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

#EJERCICIO 1

```python
header('1. Wordcloud sin procesar')
texto_min=texto.lower()
def wordcloud(texto, titulo):
  plt.figure(figsize=(20,10))
  wordcloud_crudo=WordCloud(width=800, height=500, background_color='white', collocations=False).generate(texto)
  plt.imshow(wordcloud_crudo, interpolation='bilinear')
  plt.axis("off")
  plt.title(titulo, size=15, fontweight='bold')
  plt.show()

wordcloud(texto, 'Wordcloud - Texto sin procesar (crudo)')
print(texto_min)
```

#EJERCICIO 2

```python
header('2. Minimizar + Tokenizar + Stopwords')
tokens=word_tokenize(texto_min, language='spanish')
stop_words=set(stopwords.words('spanish'))
#se agregan "además", "debe", "repetidamente", "aparecen", "enviar", "señalaron", "debe", "puedes", "detiene", "siempre")
stop_a_remover=["afirmó","comentó", "varios", "además", "debe", "repetidamente", "aparecen", "enviar", "señalaron", "debe", "puedes", "detiene", "siempre"]
for i in stop_a_remover:
  stop_words.add(i)
texto_tok=[word for word in tokens if word not in stop_words and word.isalpha()]
#se imprime el wordcloud de texto_tok
cleaned_text_tok=' '.join(texto_tok)
wordcloud(cleaned_text_tok, 'Wordcloud - Minimizar + Tokenizar + Stopwords')
print(cleaned_text_tok)
```

# EJERCICIO 3

```python
header('3. Minimizar, tokenizar por palabras,remover todos los signos (incluyendo puntuación), los stopwords, lematizar e imprimir el wordcloud.')
try:
  cleaned_text_tok=cleaned_text_tok.split()
except:
  pass
#Función para remover acentos
def sin_acentos(word):
    nfkd = unicodedata.normalize("NFD", word)
    return "".join([p for p in nfkd if not unicodedata.combining(p)])

#Función para quitar los signos
lista_tokens = []
for tok in cleaned_text_tok:
    # quitar signos (pero sobre la palabra, no sobre todo el texto)
    tok = re.sub(r'[^a-zA-Z0-9áéíóúüñÁÉÍÓÚÜÑ]', '', tok)
    # quitar acentos
    tok = sin_acentos(tok)
    if tok:  # solo agregar si no quedó vacío
        lista_tokens.append(tok)


#Se lemmatiza
lemmas = []
for token_text in lista_tokens:
    #Se procesa cada palabra como un documento de spacy
    doc = nlp(token_text)
    #Extraer el lema del primer (y único) token en el documento
    lemma = doc[0].lemma_
    lemmas.append(lemma)
print(lemmas)
#se imprime el wordcloud:
cleaned_text_tok=' '.join(lemmas)
wordcloud(cleaned_text_tok, 'Ejercicio 3 - Tokens minimizados, sin signos ni acentos y lemmatizados')
```

# EJERCICIO 4

```python
header('Ejercicio 4 - Gráfico de frecuencias')
#frecuencia de palabras con seaborn (considero que es mejor que matplotlib solo)
from collections import Counter
import pandas as pd
import seaborn as sns
conteo=Counter(lemmas)
df_palabras=pd.DataFrame(conteo.most_common(20), columns=['palabra', 'frecuencia'])
#se crea la grafica:
plt.figure(figsize=(10, 6))
sns.barplot(x='frecuencia', y='palabra', data=df_palabras, palette='viridis', hue='frecuencia')
plt.title('Frecuencia de las Palabras más Comunes')
plt.xlabel('Frecuencia')
plt.ylabel('Palabras (Lemmas)')
plt.show()
```
